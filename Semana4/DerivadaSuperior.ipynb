```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion(x):
  return np.sin(x)
```

```python
def Derivada(f,x,h=1e-4):
  return (-1/12*f(x-3*h)+1/2*f(x-2*h)-18/12*(f(x-h))+10/12*f(x)+1/4*f(x+h))/h
```

```python
x = np.linspace(0,2*np.pi,100)
y = Derivada(funcion,x)
```

```python
plt.plot(x,np.cos(x))
plt.plot(x,y,color='r')
```

```python
plt.plot(x,np.abs(np.cos(x)-y))
plt.yscale('log')
# Las fluctuaciones se deben al error de redondeo
```
