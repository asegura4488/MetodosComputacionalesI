```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as spo
```

```python
def f(x,b):
  return x**2 -b*x + 20
```

```python
x = np.linspace(-10,10,100)
y = f(x,b=-5)
```

```python
plt.plot(x,y)
```

```python
x0 = 0
result = spo.minimize(f,x0,args=(-5),options={'disp':True},method='Nelder-Mead')
```

```python
result.x
```

```python
# Problema del Jardin
```

```python
def A(p):
  return -p[0]*p[1]
```

```python
constraints = ( {'type':'eq', 'fun': lambda p: 2*p[0] + p[1] - 200} )
bounds = ((1,100),(1,100))
```

```python
p0 = np.array([10,10])
result = spo.minimize(A,p0,options={'disp':True},constraints=constraints,bounds=bounds)
```

```python
result
```

```python
result.x
```
