```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = 0.,0.
_x = np.array([x])
_y = np.array([y])
```

```python
N = 0
for i in range(20):
    theta = np.random.uniform(0,2*np.pi)
    x += 0.1*np.cos(theta)
    y += 0.1*np.sin(theta)
    _x = np.append(_x,x)
    _y = np.append(_y,y)
    N += 1
    if np.sqrt(x**2+y**2) > 1:
        print(N)
        break
    #print(x,y)
```
