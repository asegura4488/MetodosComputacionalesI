```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la funcin f(x) = x^3
def f(x):
    return x**3

# Funcin para calcular la suma de Riemann
def suma_riemann(n):
    a, b = 0, 2  # Intervalo de integracin
    delta_x = (b - a) / n
    x_i = np.linspace(a, b, n, endpoint=False)  # Puntos nodales
  #  print(x_i)
    suma = np.sum(f(x_i) * delta_x)
    return suma

# Valor exacto de la integral
def valor_exacto():
    return (2**4) / 4

suma_riemann(10)

# Valores de n
n_values = np.linspace(30, 400, 100, dtype=int)

# Clculo de la integral estimada para cada n
suma_riemann_values = np.array([suma_riemann(n) for n in n_values])

# Valor exacto de la integral
integral_exacta = valor_exacto()

# Grfico de la integral estimada y valor exacto
plt.figure(figsize=(10, 6))
plt.plot(n_values, suma_riemann_values, label='Suma de Riemann estimada', color='blue')
plt.axhline(y=integral_exacta, color='red', linestyle='--', label=f'Valor exacto: {integral_exacta}')
plt.xlabel('n (nmero de subintervalos)')
plt.ylabel('Valor de la integral estimada')
plt.title('Comparacin entre la suma de Riemann y el valor exacto')
plt.legend()
plt.grid(True)

# Mostrar el grfico
plt.show()
```
